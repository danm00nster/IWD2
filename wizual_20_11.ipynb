```python
import pandas as pd
import seaborn as sns
#!git clone https://github.com/matzim95/ML-datasets
waluty= pd.read_csv('ML-datasets/ves-usd.csv', parse_dates=True, index_col='Date', thousands=',')
```

```python
print(waluty)
waluty.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig,ax=plt.subplots(figsize=(20,15))
x=waluty.index
y=waluty.Rate
ax.plot(x,y, color='b',linestyle='--' )
ax.xaxis.set_tick_params(rotation=90)
ax.set_title('USD - Bolivar rate', color='r')
ax.set_xlabel('Date')
ax.set_ylabel('Rate')
ax.grid()
ax.annotate("disaster starts", xy=[pd.Timestamp('2019-04-15'),4000], xytext=(pd.Timestamp("2019-03-20"),4000), arrowprops={"arrowstyle":"->","color":"red"})
```
